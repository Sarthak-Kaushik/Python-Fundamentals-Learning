```python
x = 1000
type(x)
# The type of x is <class 'int'>, which indicates that x is an integer. Any number without a decimal point is considered an integer in Python.
```

```python
x = 1.5e3
type(x)
# The type of x is <class 'float'>, which indicates that x is a floating
```

```python
x = 3e+2j
type(x)
# The type of x is <class 'complex'>, which indicates that x is a complex number. Complex numbers in Python are represented as a combination of a real part and an imaginary part, where the imaginary part is denoted by 'j'. In this case, x has a real part of 0 and an imaginary part of 300 (3e+2 is equivalent to 300).
```
